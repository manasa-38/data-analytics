```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("car_price_dataset.csv")
df
```

```python
plt.figure(figsize=(15, 10))  
```

```python
Brand=pd.get_dummies(df['Brand'])
```

```python
Model=pd.get_dummies(df['Model'])
```

```python
Fuel_Type=pd.get_dummies(df['Fuel_Type'])
```

```python
Transmission=pd.get_dummies(df['Transmission'])
```

```python
df.drop(["Brand","Model","Fuel_Type","Transmission"],axis=1,inplace=True)
```

```python
df=pd.concat([df,Brand],axis=1)
df=pd.concat([df,Model],axis=1)
df=pd.concat([df,Fuel_Type],axis=1)
df=pd.concat([df,Transmission],axis=1)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df.drop('Price',axis='columns'),df['Price'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
top_10_brands = df["Brand"].value_counts()
print(top_10_brands)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
plt.pie(top_10_brands, labels=top_10_brands.index, autopct="%1.1f%%", startangle=140, colors=plt.cm.Paired.colors)
plt.title("Top 10 Car Brands")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df["Price"], bins=30, edgecolor="black", alpha=0.7)
plt.title("Distribution of Car Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=df["Year"], y=df["Price"])
plt.title("Car Price vs. Year")
plt.xlabel("Year")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Price"], kde=True, color="skyblue")
plt.title("Distribution of Car Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Engine_Size"], y=df["Price"], alpha=0.5, color="blue")
sns.regplot(x=df["Engine_Size"], y=df["Price"], scatter=False, color="red")  # Trend line
plt.title("Engine Size vs. Car Price")
plt.xlabel("Engine Size (L)")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df["Year"], y=df["Price"], marker="o", color="green")
plt.title("Car Year vs. Price Trend")
plt.xlabel("Year")
plt.ylabel("Price ($)")
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Mileage vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Mileage"], y=df["Price"], alpha=0.5, color="purple")
plt.title("Mileage vs. Car Price")
plt.xlabel("Mileage (miles)")
plt.ylabel("Price ($)")
plt.show()
```

```python
# Box plot for Owner Count vs. Price
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["Owner_Count"], y=df["Price"], palette="coolwarm")
plt.title("Owner Count vs. Car Price")
plt.xlabel("Number of Previous Owners")
plt.ylabel("Price ($)")
plt.show()
```
