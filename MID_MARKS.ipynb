# Importing the required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading the midmarks dataset using pandas

```python
df=pd.read_excel("MIDMARKS.xlsx")
df
```

```python
df['SECTION'] = df['SECTION'].fillna('SIGMA') 
df['SECTION'] = df['SECTION'].replace('', 'SIGMA')
df
```

```python
df['S.NO'] = range(1, len(df) + 1)
df
```

#  replacing all missing values in the DataFrame df with 0.

```python
df['FL'] = df['FL'].fillna(0) 
df=df.dropna()
df
```

```python
df.info()
```

```python
columns_to_convert = ["DV", "M-II", "PP", "BEEE", "FL", "FIMS"]


df[columns_to_convert] = df[columns_to_convert].apply(lambda col: pd.to_numeric(col, errors="coerce"))


df[columns_to_convert] = df[columns_to_convert].fillna(0)


df[columns_to_convert] = df[columns_to_convert].astype(int)


print(df.dtypes)
```

# Creating a new column Total , which is the sum of all subject scores

```python
df['Total'] = df['DV'] + df['M-II'] + df['PP'] + df['BEEE'] + df['FL'] + df['FIMS']
df
```

```python
df.sort_values("Total", ascending=True,inplace=True)
fdtail=df.tail(10)
fdtail
```

```python
#Distribution of scores for each subject
plt.figure(figsize=(15, 10))
for i, column in enumerate(columns_to_analyze, 1):
    plt.subplot(2, 3, i)
    for section in sections:
        section_data = df[df['SECTION'] == section]
        plt.hist(section_data[column], bins=10, alpha=0.5, label=section, edgecolor='black')
    plt.title(f'Score Distribution: {column}')
    plt.xlabel('Scores')
    plt.ylabel('Frequency')
    plt.legend(title='Section')
plt.tight_layout()
plt.show()
```
